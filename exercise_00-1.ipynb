### Creating your own functions

Another often helpful feature in Python is defining your own functions. A function contains an algorithm that you specify when creating it. Afterwards, the function can be called and used by its name and arguments:

Functions can manipulate values, print information to the console (with <code>print</code>), and/or return values. To return values, use the <code>return</code> statement:

```python
def potenz(a, b):
    result = a**b
    return result

def minus(c, d, e):
    f = potenz(c,d)
    result = f - e
    return result
        
ergebnis = minus(5, 3, 2) 
print(ergebnis)
```

<code>def potenz(a, b):</code> means that a new function is defined (<code>def</code>) with a name (<code>potenz</code>), which expects arguments <code>(a, b)</code> when called.

Inside the function, an algorithm is executed (<code>result = a**b</code>) and returned (<code>return result</code>).

The function can be used, for example, with <code>result = potenz(5, 3)</code>. The return value is stored in a variable and can then be printed using <code>print</code>. This also shows that one function can be called inside another.

### Class structure

You may already know the concept of classes from an earlier programming course. Their main purpose is to save code and time. A class in Python describes a <b>type</b> of thing. A simple mechanical example is the class "rod" (German: Stab). Rods generally share the same kinds of properties, for example a length, a Young’s modulus, and a cross-sectional area. The basic structure is created in the following cell:

```python
class Stab:
    def __init__(self, L_mm, E_MPa, A_mm2):
        self.L_mm = L_mm
        self.E_MPa = E_MPa
        self.A_mm2 = A_mm2
```

<code>class Stab:</code> means that the indented code that follows defines the class Stab.

<code>def __init__(self, L_mm, E_MPa, A_mm2):</code> defines a method to create a new instance (i.e., a new object) of the class Stab. You must pass 3 arguments, which can then be accessed inside the class via the names <code>L_mm</code>, <code>E_MPa</code>, <code>A_mm2</code>.

<code>self.L_mm = L_mm</code>,

<code>self.E_MPa = E_MPa</code>,

<code>self.A_mm2 = A_mm2</code> assigns the arguments to the newly created object. You can still access these values after the method finishes; they become properties of the object.

Now you can create instances of the class Stab. You can use the code in the following cell:

```python
Stab_a = Stab(100, 80000, 25)
```

The code cell above creates a new object of the class Stab. By your definition, the rod has a length of 100 mm, a Young’s modulus of 80,000 MPa, and a cross-sectional area of 25 mm². When you run the cell, the method <code>__init__()</code> is called in the background and creates the object; all expressions inside the method are evaluated.



For example, you can access the rod length using <code>Stab_a.L_mm</code>.

```python
Stab_a.L_mm
```

To make this class actually useful, you can define a method that computes the change in length of the rod. The following formula applies:

$\Delta l = \frac{F l}{EA}$ 

Change in length: $\Delta l$

Normal force: $F$

Young’s modulus: $E$

Cross-sectional area: $A$

```python
class Stab:
    def __init__(self, L_mm, E_MPa, A_mm2):
        self.L_mm = L_mm
        self.E_MPa = E_MPa
        self.A_mm2 = A_mm2
        
    def berechne_laengenaenderung(self, F_N):
        deltaL_mm = self.L_mm * F_N / (self.E_MPa * self.A_mm2)
        return(deltaL_mm)
```

The method <code>berechne_laengenaenderung()</code> is now available for every object of the class Stab. To distinguish it from other functions and to indicate that it is a class method, <code>self</code> is passed as the first argument. You do not provide <code>self</code> explicitly when calling the method.

Inside the method, you access the corresponding instance variables via expressions like <code>self.L_mm</code>. This always refers to the properties of the <b>current</b> object.

The method berechne_laengenaenderung requires the additional argument <code>F_N</code> and computes the length change for the rod it is applied to. For example, you can compare the elongations of two rods with the same load and the same dimensions:

```python
Stab_PP = Stab(1000, 1600, 25)
Stab_Stahl = Stab(1000, 210000, 25)

print(Stab_PP.berechne_laengenaenderung(1000))
print(Stab_Stahl.berechne_laengenaenderung(1000))
```

### Matplotlib

```python
import matplotlib.pyplot as plt
```

The matplotlib library can be used to visualize data, which is often helpful for interpreting it.

Here we give a short introduction to this library, which was imported in the previous code cell. If you have not done so yet, run the previous cell once.

Let’s start by generating data that we want to plot — for example, sine and cosine.

<code>np.linspace(0, 2*np.pi, 100)</code> creates an array with 100 evenly spaced values between 0 and $2\pi$.

After executing the cell below, <code>y1</code> and <code>y2</code> are also arrays containing the function values.

```python
x = np.linspace(0, 2* np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x) 
```

Now we want to visualize the data (i.e., plot it)

<code>plt.plot(x,y)</code> plots the x values and y values into a diagram.

With <code>label=</code> you set the name that the curve should have in the legend.

With <code>color=</code> you can choose a color. If you do not specify one, Python will pick one automatically.

Additionally, you can label the axes using <code>plt.xlabel()</code> and <code>plt.ylabel()</code>.

With <code>plt.legend()</code> you tell Python to display the labels in the plot.

With <code>plt.show()</code> you tell Python that you are done plotting and want to display the figure. If you want to show two curves in separate figures, you would need to call <code>plt.show()</code> in between.

```python
plt.plot(x, y1, label='sin(x)', color='blue')
#plt.legend()
#plt.show()
plt.plot(x, y2, label='cos(x)', color='red')
plt.xlabel('x coordinate (abscissa)')
plt.ylabel('y coordinate (ordinate)')
plt.legend()
plt.show()
```
